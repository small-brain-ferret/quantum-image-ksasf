# Grayscale Pixel Intensity and Colorbars

This notebook demonstrates how pixel intensity values affect grayscale images, using both horizontal and vertical colorbars.

```python
# Import Required Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate Grayscale Intensity Data
# Create a 2D gradient from 0 (black) to 255 (white)
width, height = 256, 32
horizontal_gradient = np.tile(np.linspace(0, 255, width), (height, 1))
vertical_gradient = np.tile(np.linspace(0, 255, height), (width, 1)).T
```

```python
# Create Horizontal Colorbar
plt.figure(figsize=(8, 2))
plt.imshow(horizontal_gradient, cmap='gray', aspect='auto', vmin=0, vmax=255)
cbar = plt.colorbar(orientation='horizontal', pad=0.2)
cbar.set_label('Pixel Intensity', fontsize=14)
plt.title('Horizontal Colorbar: Grayscale Pixel Intensity', fontsize=16)
plt.axis('off')
plt.show()
```

```python
# Create Vertical Colorbar
plt.figure(figsize=(2, 6))
plt.imshow(vertical_gradient, cmap='gray', aspect='auto', vmin=0, vmax=255)
cbar = plt.colorbar(orientation='vertical', pad=0.2)
cbar.set_label('Pixel Intensity', fontsize=14)
plt.title('Vertical Colorbar: Grayscale Pixel Intensity', fontsize=16)
plt.axis('off')
plt.show()
```
